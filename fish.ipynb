```python
import numpy as np
import matplotlib.pyplot as plt

fish = plt.imread('Aquarium_fish.jpg').astype(float)
fish = (fish - fish.min()) / (fish.max() - fish.min())
fish.shape

fig, ax = plt.subplots(figsize=(10,15))
image= ax.imshow(fish)


fig, ax = plt.subplots(figsize=(10,15))
image = ax.imshow(fish[:, :, 0], cmap=plt.cm.Reds)

fish_bw = fish.mean(axis=2)

fig, ax = plt.subplots(figsize=(10,15))
image = ax.imshow(fish_bw, cmap=plt.cm.binary)



import matplotlib.animation as animation

fig, ax = plt.subplots(figsize=(10,15))
image = ax.imshow(fish)

def update(frame):
    shift = frame % fish.shape[1]  # Calculate the shift amount
    fish_shifted = np.concatenate([fish[:, fish.shape[1]-shift:], fish[:, :fish.shape[1]-shift]], axis=1)
    image.set_data(fish_shifted)
    return image

ani = animation.FuncAnimation(fig, update, frames=fish.shape[1], interval=30)
plt.show()

from PIL import Image, ImageDraw, ImageOps


def make_circular(im):
    offset = (im.width - im.height) // 2
    im = im.crop((offset, 0, im.height + offset, im.height))

    mask = Image.new("L", im.size, 0)
    draw = ImageDraw.Draw(mask)
    draw.ellipse((0, 0) + im.size, fill=255)

    out = ImageOps.fit(im, mask.size, centering=(0.5, 0.5))
    out.putalpha(mask)
    return out


def round_edges(im, radius):
    mask = Image.new("L", (radius * 2, radius * 2), 0)
    draw = ImageDraw.Draw(mask)
    draw.ellipse((0, 0, radius * 2, radius * 2), fill=255)

    alpha = Image.new("L", im.size, 255)
    w, h = im.size

    alpha.paste(mask.crop((0, 0, radius, radius)), box=(0, 0))
    alpha.paste(mask.crop((0, radius, radius, radius * 2)), box=(0, h - radius))
    alpha.paste(mask.crop((radius, 0, radius * 2, radius)), box=(w - radius, 0))
    alpha.paste(mask.crop((radius, radius, radius * 2, radius * 2)), box=(w - radius, h - radius))

    im.putalpha(alpha)
    return im


if __name__ == "__main__":
    with Image.open("Aquarium_fish.jpg") as im:
        im = im.convert("RGBA")

        circular_im = make_circular(im)
        circular_im.save("Circular fish.png")

        rounded_im = round_edges(im, 100)
        rounded_im.save("Rounded fish.png")

  

        

import numpy as np
import cv2

# Load the image
img = cv2.imread("Aquarium_fish.jpg")


#if img is not None:
    #print('variable is not None')
    #print(img.shape)
#else:
    #print('variable is None')

# Create a black mask
mask = np.zeros(img.shape[:2], dtype=np.uint8)

# Define the center point and radius of the circular window
center = (1800, 1000)
radius = 1000

# Draw a white, filled circle on the mask
cv2.circle(mask, center, radius, (255, 255, 255), -1)

# Apply the mask to the image
img = cv2.bitwise_and(img, img, mask=mask)

# Save the resulting image
cv2.imwrite("masked_fish.jpg", img)


#from PIL import Image

#img_PIL = Image.open("masked_fish.jpg")
#img_PIL.show()

cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()


import cv2
import numpy as np


def gammaCorrection(src, gamma):
    invGamma = 1 / gamma

    table = [((i / 255) ** invGamma) * 255 for i in range(256)]
    table = np.array(table, np.uint8)

    return cv2.LUT(src, table)


img = cv2.imread('Aquarium_fish.jpg')
gammaImg = gammaCorrection(img, 2.2)
imS = cv2.resize(img, (960, 540))                # Resize image

cv2.imshow('Original image', img)
cv2.imshow('Gamma corrected image', gammaImg)
cv2.waitKey(0)
cv2.destroyAllWindows()


import cv2

img = cv2.imread('Aquarium_fish.jpg', 0)

# Plot the histogram using Matplotlib
plt.hist(img.ravel(), bins=256, range=[0,256])
plt.xlabel('Gray level')
plt.ylabel('Frequency')
plt.title('Histogram of gray levels')
plt.show()

fish = plt.imread('Aquarium_fish.jpg')

# Plot the histograms of the RGB levels
fig, axs = plt.subplots(nrows=3, figsize=(10,15))
for ax, channel, color in zip(axs, ['R', 'G', 'B'], ['red', 'green', 'blue']):
    ax.hist(fish[..., 0].ravel(), bins=256, color=color)
    ax.set_xlabel('Pixel intensity')
    ax.set_ylabel('Frequency')
    ax.set_title(f'Histogram of {channel} channel')
plt.tight_layout()
plt.show()

fish = plt.imread('Aquarium_fish.jpg')
fish_bw = fish.mean(axis=2)

# Define the gamma correction values
gammas = np.linspace(0.6, 1.4, 9)

# Plot the subplots with the enhanced grayscale fish
fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(10,15))
for ax, gamma in zip(axs.ravel(), gammas):
    enhanced_fish = np.uint8(255 * (fish_bw / 255)**gamma)
    ax.imshow(enhanced_fish, cmap=plt.cm.binary)
    ax.axis('off')
    ax.set_title(f'Gamma = {gamma:.2f}')
plt.tight_layout()
plt.show()


```
