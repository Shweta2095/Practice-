```python
import csv
import pandas as pd
from typing import Dict


class Profiler:
    def __init__(self, sequence):
        self.seq = sequence

    def get_sequence(self):
        return self.seq

    def get_length(self):
        return len(self.seq)

    def is_valid(self):
        return all(nucleotide in 'ACGT' for nucleotide in self.seq)

    def longest_run(self, subseq: str) -> int:

       count = 0
       max_count = 0
       i = 0
       while i < len(self.seq) - len(subseq) + 1:
        if self.seq[i:i+len(subseq)] == subseq:
            count += 1
            i += len(subseq)
            if count > max_count:
                max_count = count
        else:
            i += 1
            count = 0
       return max_count
   
    

    #def match_suspect(self, suspect_name: str, dna_fpr_file: str) -> bool:
     #   dna_fpr_file = '/content/data.csv'

      #  try:
       #  with open(dna_fpr_file) as f:
        #    reader = csv.reader(f)
          #  header = next(reader)
         #   suspect_index = header.index(suspect_name)
        #except ValueError:
         #   return False
        #return True

    def match_suspect(self, suspect_name: str, dna_fpr: dict):
        for subseq, required_count in dna_fpr.items():
            if self.longest_run(subseq) < required_count:
              return False
        return True


      
```
